##filtro Private room   “host_acceptance_rate vs host_response_rate”



```python
california[["room_type","source"]]
```

```python
filtro1 = california[california["room_type"] == "Private room"]
filtro1
```

```python
filtro1 = filtro1.replace('sin registro', pd.np.nan)
```

```python
filtro1 = filtro1.dropna(subset=["host_acceptance_rate","host_response_rate"])
```

```python
filtro1[['host_acceptance_rate',"host_response_rate"]] = filtro1[['host_acceptance_rate',"host_response_rate"]].replace('%', '', regex=True).astype('float') / 100
```

```python
#Cargar archivo csv desde seaborn
df=sns.load_dataset(name='car_crashes')
df.head(5)
```

```python
#Imprimimos el scatter plot entre la variable dependiente (total) e independiente (alchool) 
#para observar el comportamiento en su dispersión 
from turtle import color
sns.scatterplot(x='host_acceptance_rate', y='host_response_rate', color="blue", data=filtro1)
#sns.scatterplot(X='speeding', y='total', color="red", data=df)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
#Vars_Indep= df[['alcohol', 'speeding']]
Vars_Indep= filtro1[['host_acceptance_rate']]
Var_Dep= filtro1['host_response_rate']
```

```python
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

**Modelo matemático: y= 2.0325063x + 5.857776154826299**

---




```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinación
model.score(Vars_Indep,Var_Dep)
```

```python
#Predecimos los valores de total de accidentes a partir de la variable "alcohol"
#y_pred= model.predict(X=df[['alcohol', 'speeding']])
y_pred= model.predict(X=filtro1[['host_acceptance_rate']])
y_pred
```

```python
#Insertamos la columna de predicciones en el DataFrame
filtro1.insert(0, 'Predicciones', y_pred)
filtro1
```

```python
#Visualizamos la gráfica comparativa entre el total real y el total predecido

sns.scatterplot(x='host_acceptance_rate', y='host_response_rate', color="blue", data=filtro1)
sns.scatterplot(x='host_acceptance_rate', y='Predicciones', color="red", data=filtro1)
#sns.lineplot(x='alcohol', y='Predicciones', color="red", data=df)
```

```python
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
